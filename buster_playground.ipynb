```python
motor_comms = MotorCommunicator("/dev/tty.usbmodem58FA0833071")

def dict_to_str(d):
    return"{" + ", ".join([f"{k}: {v}" for k, v in d.items()]) + "}"

filename = "test_positions.txt"
with open(filename, "w") as f:
    f.write("[")

try:
    while True:
        input("Press Enter to record current positions, or Ctrl-C to exit...")
        positions = motor_comms.get_motor_positions(joints.ALL_LEG_JOINT_IDS)
        print(positions)
        with open(filename, "a") as f:
            positions = motor_comms.get_motor_positions(joints.ALL_LEG_JOINT_IDS)
            f.write(dict_to_str(positions) + ",\n")
except KeyboardInterrupt:
    pass

finally:
    with open(filename, "a") as f:
        f.write("]")

```
